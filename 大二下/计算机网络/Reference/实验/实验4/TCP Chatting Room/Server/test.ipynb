```python
import Ipynb_importer
from config import *
import socket
import os

def test():  # 测试服务器连接和数据收发
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect(("127.0.0.1", 8090))
    
    sending = False
    
    while True:
        message = input("Input the message to send: ")
        if message == "/logout": 
            break
        elif message == "sending": 
            sending = True
            continue
            
        client_socket.send(message.encode("utf-8"))  # 向服务器发送消息
        recv_data = client_socket.recv(512)  # 接收服务器发来的消息
        if not sending:
            print(recv_data.decode("utf-8"))
        else:
            print(recv_data)
            print(recv_data.decode("utf-8"))
            filename = "received_small.txt"
            filepath = os.path.join(SAVE_PATH, filename)
            file_size = 56
            recv_size = 0
            file = open(filepath, "wb")
            sending = True
            
            while sending:
                data = client_socket.recv(512)  # 接收服务器发来的消息
                data = data.decode("utf-8")
                if recv_size + SEND_CHUNK >= file_size:  # 最后一批
                    print("Received a chunk of %d Bytes." % (file_size - recv_size))
                    sending = False
                else:  # 非最后一批
                    print("Received a chunk of 64 Bytes.")
                    recv_size += SEND_CHUNK
                
                try:
                    file.write(bytes(data, encoding = "utf-8"))
                except:
                    continue
            
            file.close()
            print("Received the file %s (%d Bytes)." % (filename, file_size))

    client_socket.close()  # 关闭客户端套接字
    
if __name__ == "__main__":
    test()
```
